# Octave-Sandbox
## zum Skript »Einführung in Algorithmik und Programmieren mit Octave/MATLAB«

Willkommen in Ihrer interaktiven Programmierumgebung - **ohne Installation und ohne Anmeldung**.  
Öffnen Sie parallel das PDF‑Skript und schreiben Sie Ihren Code in die grauen Zellen oder tippen Sie die Code‑Beispiele ab, um sie direkt auszuführen.

---

### Kurzanleitung

- **Code ausführen:** Klicken Sie in die entsprechende Codezelle und drücken Sie `Shift` + `Enter` (oder den ▶ Knopf in der Werkzeugleiste).
- **Eigene Funktionen schreiben:** Schreiben Sie eine Funktion (`function ... end`) in eine eigene Codezelle und führen Sie diese aus. Rufen Sie die Funktion dazu **in einer neuen Zelle** auf, nicht in derselben.
- **Selbst abtippen:** Bitte kopieren Sie den Code nicht aus dem PDF (versteckte fehlerhafte Zeichen), sondern tippen Sie ihn ab. So lernen Sie effektiver.
- **Session‑Reset:** Bei längerer Inaktivität können Variablen verloren gehen. Führen Sie dann die nötigen Zellen einfach erneut aus.
- **Zellen ergänzen:** Sie können bei Bedarf beliebig viele weitere Code‑ oder Textzellen hinzufügen.
